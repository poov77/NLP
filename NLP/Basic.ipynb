```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# 1. Training Data (Text samples and their correct categories)
# 1 = Positive sentiment, 0 = Negative sentiment
training_sentences = [
    "I absolutely love this product, it is amazing",
    "This is the best app I have ever used",
    "Fantastic experience, highly recommended",
    "This is terrible and worst experience ever",
    "I hate this, it is completely useless",
    "Very bad quality, do not buy this product"
]
labels = [1, 1, 1, 0, 0, 0]  # 3 positive, 3 negative

# 2. Vectorization: Turning Text into Math
# CountVectorizer strips punctuation and converts text to lowercase automatically.
vectorizer = CountVectorizer()

# The model 'learns' all unique words and maps them to numbers
X_train = vectorizer.fit_transform(training_sentences)

# 3. Choose the Classifier 
# Multinomial Naive Bayes is the standard classic algorithm for text counting.
model = MultinomialNB()
model.fit(X_train, labels)

# 4. Test it with totally new sentences the model has never seen!
test_sentences = [
    "I love it, absolutely fantastic!",
    "This app is terrible and bad"
]

# Convert the test sentences using the same rules learned during training
X_test = vectorizer.transform(test_sentences)

# Make the predictions
predictions = model.predict(X_test)

# 5. Print out the results nicely
for sentence, prediction in zip(test_sentences, predictions):
    sentiment = "Positive" if prediction == 1 else "Negative"
    print(f"Sentence: '{sentence}' -> Predicted Sentiment: {sentiment}")
```
