```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain.chains import RetrievalQA   # ✅ available in 1.3.11
from langchain_core.prompts import ChatPromptTemplate

# 1. Load the PDF file
loader = PyPDFLoader("NLP/History of Dharmapuri District.pdf")
docs = loader.load()

# 2. Chop the PDF into smaller chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = text_splitter.split_documents(docs)

# 3. Embed the text chunks and save them to a vector database
embeddings = OpenAIEmbeddings()
vector_store = Chroma.from_documents(chunks, embeddings)

# 4. Turn the database into a retriever tool
retriever = vector_store.as_retriever(search_kwargs={"k": 3})

# 5. Connect to an LLM
llm = ChatOpenAI(model="gpt-4")

# 6. Build a RetrievalQA chain (works in 1.3.11)
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    chain_type="stuff"
)

# Try asking it a question!
response = qa_chain.run("What is the main summary of this document?")
print(response)
```

```python
from langchain_pdf import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain.chains.retrieval import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate

# 1. Load the PDF file
loader = PyPDFLoader("NLP/History of Dharmapuri District.pdf")
docs = loader.load()

# 2. Split into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = text_splitter.split_documents(docs)

# 3. Embed + store
embeddings = OpenAIEmbeddings()
vector_store = Chroma.from_documents(chunks, embeddings)

# 4. Retriever
retriever = vector_store.as_retriever(search_kwargs={"k": 3})

# 5. LLM + prompt
llm = ChatOpenAI(model="gpt-4o-mini")
system_prompt = (
    "You are an assistant for question-answering tasks. "
    "Use the following pieces of retrieved context to answer "
    "the question. If you don't know the answer, say that you don't know.\n\n"
    "{context}"
)
prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("human", "{input}"),
])

# 6. Chain
question_answer_chain = create_stuff_documents_chain(llm, prompt)
rag_chain = create_retrieval_chain(retriever, question_answer_chain)

# Test
response = rag_chain.invoke({"input": "What is the main summary of this document?"})
print(response["answer"])
```
